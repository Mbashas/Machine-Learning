```python
### BEGIN SOLUTION
# Using Matplotlib's plotting functionality
ax = plt.axes()
ax.hist(data.Age, bins=25);

ax.set(xlabel='HighBP', 
       ylabel='Frequency',
       title='Distribution of Petal Lengths');
```

```python
# Select specific columns to create a new dataset
data1 = ['Age', 'BMI', 'Smoker', 'Sex','Education']
data1 = data[data1]
data1.head()
```

```python
### BEGIN SOLUTION
# First we have to reshape the data so there is 
# only a single measurement in each column

plot_data = (data1
             .set_index('Sex')
             .stack()
             .to_frame()
             .reset_index()
             .rename(columns={0:'size', 'level_1':'measurement'})
            )

plot_data.head()
### END SOLUTION
```

```python
### BEGIN SOLUTION
# Now plot the dataframe from above using Seaborn
import seaborn as sns
sns.set_style('white')
sns.set_context('notebook')
sns.set_palette('dark')

f = plt.figure(figsize=(6,4))
sns.boxplot(x='measurement', y='size', 
            hue='Sex', data=plot_data);
### END SOLUTION
```

```python
### BEGIN SOLUTION
sns.set_context('talk')
sns.pairplot(data1, hue='Sex');
### END SOLUTION
```

```python
# Drop non-numeric columns
import numpy as np


numeric_cols = data1.select_dtypes(include=[np.number])

plt.figure(figsize=(18,18))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='RdYlGn')

plt.show()
```
